```python

%pip install pandas
%pip install seaborn
%pip install matplotlib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python

csv_path = "Running_Sims/CSV data/run_2026-02-20T14-23-09/kbuffer_one_run_2026-02-20T14-23-09.csv"
df = pd.read_csv(csv_path)
```

```python

df["Generation"] = pd.to_numeric(df["Generation"], errors="coerce")
df["population"] = pd.to_numeric(df["population"], errors="coerce")
df["Rep"] = df["Rep"].astype(str)  # color by replicate

# Plot one line per replicate
plt.figure(figsize=(10,6))
sns.lineplot(
    data=df,
    x="Generation",
    y="population",
    hue="Rep",
    marker="o"
)
plt.title("Population across generations by replicate")
plt.xlabel("Generation")
plt.ylabel("Population")
plt.legend(title="Rep", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
